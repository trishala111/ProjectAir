```python
import pandas as pd
import plotly.express as px
import plotly
```

```python
df = pd.read_csv("waterquality.csv")
df["Time"] = pd.to_datetime(df["Time"])
df.dtypes
```

```python
df = df.sort_values(by = "Time")
```

```python
df["Time"] = df["Time"].astype(str)
```

```python
df[["latitude","longitude"]] = df["AOI"].str.split("," , expand = True)
df["latitude"] = df["latitude"].astype(float)
df["longitude"] = df["longitude"].astype(float)
df.dtypes
```

```python
df_dropped = df.dropna(axis = "index" ,subset = ["Measurement Value"])
df_refined = df_dropped.dropna(axis = "columns" , how = "all")
df_refined
```

```python
    map_fig = px.density_mapbox(df_refined, lat = "latitude", lon = "longitude" ,z= "Measurement Value", 
                           mapbox_style="carto-darkmatter",
                                animation_frame = "Time",
                                hover_name = "Region",
                                hover_data = ["Country","Indicator Name","Method","EO Sensor"],
                             radius = 10,
                                opacity = 1,
                           zoom = 2)
    map_fig.show()
```

```python
plotly.offline.plot(map_fig, filename = "water1.html" , auto_open = True)
```
